```python
import matplotlib.pyplot as plt
```

```python
Adam={'Learning Rates':['2e-5','3e-5','5e-5'],'Exact Match Score':[0.75,0.74,0.75]}
RMSprop={'Learning Rates':['2e-5','3e-5','5e-5'],'Exact Match Score':[0.76,0.73,0.75]}
Adagrad={'Learning Rates':['2e-5','3e-5','5e-5'],'Exact Match Score':[0.6,0.65,0.55]}
Adamax={'Learning Rates':['2e-5','3e-5','5e-5'],'Exact Match Score':[0.76,0.78,0.78]}
Nadams={'Learning Rates':['2e-5','3e-5','5e-5'],'Exact Match Score':[0.77,0.78,0.78]}
```

```python
name=[Adam,RMSprop,Adagrad,Adamax,Nadams]
s=["Adam","RMSprop","Adagrad","Adamax","Nadams"]
c=['r','b','g','pink','purple']
for i in range(5):
    plt.plot(name[i]['Learning Rates'],name[i]['Exact Match Score'],linestyle='--' ,color=c[i],Label=s[i])
    

    
plt.legend(fontsize=13)                                     # Display the legend
plt.xlabel('Learning Rates', fontsize=13)                                # Label for x data
plt.ylabel('Exact Match Score', fontsize=13)     
plt.gca().legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()     
```
